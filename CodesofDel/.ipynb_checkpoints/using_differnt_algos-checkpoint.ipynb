```python
from sklearn.ensemble import RandomForestClassifier
```

```python
features = [
    [150,0],
    [140,1],
    [160,0],
    [120,1],
]
```

```python
labels = ["apple","apple","orange","orange"]
```

```python
clf = RandomForestClassifier()
```

```python
predict = clf.fit(features, labels)
```

```python
print(predict)
```

```python
clf.predict([[150,0]])
```

```python
prediction = clf.predict([[120,1]])
```

```python
print(f"The fruit is {prediction}")
```
