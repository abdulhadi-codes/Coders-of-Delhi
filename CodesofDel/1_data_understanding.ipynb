```python
import json
```

```python
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)

    return data
        
        
```

```python
#cleaning the data 
def display_users(data):
    print("The data in Collection")
    for user in data['users']:
        print(f"ID {user['id']} - {user['name']} and friends of {user['friends'] } and the name of the pages {user['liked_pages']}")
    print("The pages are")
    for pages in data['pages']:
        print(f"{pages['id']} and the {pages['name']}")

display_users(data)
    
    
        
        
    
    
```
